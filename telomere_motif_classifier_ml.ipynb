## CREATING DATAFRAME

```python
df = pd.read_csv("demo_data5.csv")
x = df.values
x = np.delete(x, 2, axis=1)
y = df['Label'].values
print(x)
print(y)
```

## SPLITTING OF DATA INTO TRAINING AND TEST DATA

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.3,random_state=3)

print(X_test)
print(Y_train)
```

## FITING OF MODEL

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
myleo = leo()
y_train_new = myleo.fit_transform(Y_train)
y_test_new = myleo.fit_transform(Y_test)
x_train_new1 = myleo.fit_transform(X_train[:,0])
x_train_new2 = myleo.fit_transform(X_train[:,1])
x_test_new1 = myleo.fit_transform(X_test[:,0])
x_test_new2 = myleo.fit_transform(X_test[:,1])
x_train_new = []
x_test_new = []
for i in range(len(x_train_new1)):
    x_train_new.append([x_train_new1[i], x_train_new2[i]])
for j in range(len(x_test_new1)):
    x_test_new.append([x_test_new1[j], x_test_new2[j]])

sc = StandardScaler()
x_train_new = sc.fit_transform(x_train_new)
x_test_new = sc.transform(x_test_new)
print(x_train_new)
print(x_test_new)
cls = SVC()
cls.fit(x_train_new, y_train_new)

```

## SVM

```python
y_pred = cls.predict(x_test_new)
print(y_pred)
print(y_test_new)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test_new, y_pred)
print(cm)
accuracy_score(y_test_new, y_pred)

```

## RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train_new, y_train_new)
yp = rf.predict(x_test_new)
print(yp)
print(y_test_new)
c = confusion_matrix(y_test_new, yp)
print(c)
accuracy_score(y_test_new, yp)
```

## MULTI LAYER PERCEPTRON

```python
from sklearn.neural_network import MLPClassifier
mc = MLPClassifier()
mc.fit(x_train_new, y_train_new)
ypr = mc.predict(x_test_new)
print(ypr)
print(y_test_new)
co = confusion_matrix(y_test_new, ypr)
print(co)
accuracy_score(y_test_new, ypr)
```

## K-NEAREST-NEIGHBOUR

```python
knn = KNeighborsClassifier()
knn.fit(x_train_new, y_train_new)
ky = knn.predict(x_test_new)
print(ky)
print(y_test_new)
conf = confusion_matrix(y_test_new, ky)
print(conf)
accuracy_score(y_test_new, ky)
```
