```python
!pip install Bio
from Bio import SeqIO
import matplotlib.pyplot as pt

def count(list1, l, r):
    c = 0
    for x in list1:
        if x>= l and x<= r:
            c+= 1
    return c

def dummy_fun(seq,ptt):
    n = len(seq)
    lst = []
    clst = []
    #ptt_lst = ['TTAGGG', 'GGGATT', 'AATCCC', 'CCCTAA']
    ptt_lst = [ptt]
    for i in range(0, n):
        m = seq[i:i+len(ptt)]
        if(m in ptt_lst):
            lst.append(i)
    low = 0
    high = 1000
    while high<n:
        c = count(lst, low, high)
        clst.append(c)
        low += 1000
        high += 1000

    return clst

seq = SeqIO.read("chr22.fa", "fasta")
final = seq.seq
usr_ptt = input("Enter pattern: ").upper()
value = dummy_fun(final, usr_ptt)
My_list = [range(0, len(final)//1000, 1)]
pt.title("Occurences of {} in Chromosome 22 (Human)".format(usr_ptt))
pt.xlabel("Repeats per 1000 bases")
pt.ylabel("No of repeats")
pt.scatter(My_list, value, s=5)
pt.show()
```
